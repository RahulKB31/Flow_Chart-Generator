```python
import matplotlib.pyplot as plt
import networkx as nx

# -----------------------------------------------------------
# High-Level Data Flow – Particle Counter Dashboard
# -----------------------------------------------------------

# Create a directed graph
G = nx.DiGraph()

# Define nodes and labels
nodes = {
    "A": "Start: Data Collection\n(CSV Files – Particle Counter Readings)",
    "B": "Database Connection\n(PostgreSQL Tables Import)",
    "C": "Data Preprocessing\n(Cleaning, Transformation, Integration)",
    "D": "Data Modeling\n(Relationships + DAX Measures)",
    "E": "Visualization\n(Line Charts, KPIs, Analytical Table)",
    "F": "Testing & Deployment\n(Power BI Service, Scheduled Refresh)",
    "G": "Maintenance & Optimization\n(Performance, Data Updates)"
}

# Add nodes to the graph
for node, label in nodes.items():
    G.add_node(node, label=label)

# Define directional edges (top-down flow)
edges = [
    ("A", "B"),
    ("B", "C"),
    ("C", "D"),
    ("D", "E"),
    ("E", "F"),
    ("F", "G")
]
G.add_edges_from(edges)

# Define node positions (vertical top-down layout)
pos = {
    "A": (0, 6),
    "B": (0, 5),
    "C": (0, 4),
    "D": (0, 3),
    "E": (0, 2),
    "F": (0, 1),
    "G": (0, 0)
}

# -----------------------------------------------------------
# Plot the Directed Graph
# -----------------------------------------------------------
plt.figure(figsize=(10, 11))
nx.draw(
    G, pos,
    with_labels=False,
    arrows=True,
    node_size=4200,
    node_color="#AED6F1",
    edgecolors='black',
    linewidths=1.2,
    arrowsize=18,
    connectionstyle="arc3,rad=0.0"
)

# Draw node labels (centered and wrapped)
for node, (x, y) in pos.items():
    plt.text(
        x, y,
        nodes[node],
        fontsize=10,
        ha='center',
        va='center',
        wrap=True,
        fontweight='semibold'
    )

# Add title and layout settings
plt.title("Particle Counter Dashboard – High-Level Solution Flow", fontsize=14, fontweight='bold')
plt.axis('off')
plt.tight_layout()

# Save and display the flowchart
plt.savefig("particle_counter_dashboard_flowchart.png", dpi=300, bbox_inches='tight')
plt.show()
```
